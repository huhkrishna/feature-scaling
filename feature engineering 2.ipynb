```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.iloc[:,2:4],df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
```

```python
ms.fit(X_train)
```

```python
xtrainscaled=ms.transform(X_train)
xtestscaled=ms.transform(X_test)
```

```python
xtrainscaled=pd.DataFrame(xtrainscaled,columns=X_train.columns)
xtestscaled=pd.DataFrame(xtestscaled,columns=X_test.columns)
```

```python

np.round(xtrainscaled.describe(),1)
```

```python
np.round(X_train.describe(),1)
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title('Before')
ax2.scatter(xtrainscaled['Age'],xtrainscaled['EstimatedSalary'],color='red')
ax2.set_title('after')
plt.show()
```
